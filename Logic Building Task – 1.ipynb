```python
#1. User Login Check
#Checking the Credentials
stored_us = "admin"
stored_pass = "1234"
username = input("Enter the Username : ")
password = input("Enter the Password : ")
#Login Check
if((username == stored_us) and (password == stored_pass)):
    print("Login Successful")
else :
    print("Invalid Credentials")
```

```python
#2. Pass/Fail Analyser
marks = [45,78,90,33,60]
pass_count = 0
fail_count = 0
for i in marks:
    if(i >= 50):
        pass_count+=1
    else:
        fail_count+=1
print("Total No.of Students : ",len(marks))
print("No.of Students who have Passed : ",pass_count)
print("No.of Students who have Failed : ",fail_count)
```

```python
#3. Simple Data Cleaner
#Original List
names = [" Alice "," bob "," CHARLIE "]
cleaned_list = []
cleaned_list = [name.strip().lower() for name in names]
print("Cleaned List = ",cleaned_list)
```

```python
#4. Message length Analyzer
messages = ["Hi","Welcome to the platform","OK"]
for message in messages:
    length = len(message)
    print("Message : ",message)
    print("Length : ",length)
    if(length > 10):
        print("Status : Flag Message")
    else :
        print("Status : Not a Flag Message")
```

```python
#5. Error Message Detector
logs = ["INFO","ERROR","WARNING","ERROR"]
error_cnt = 0
for log in logs:
    if(log == "ERROR"):
        error_cnt+=1
print("Error Count = ",error_cnt)
```
