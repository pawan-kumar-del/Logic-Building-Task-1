```python
# Problem 1
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Highest Score
high_score = max(employees.values())

# Matching employees with that score
top_performers = [name for name, score in employees.items() if score == high_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {high_score})")
```

```python
# Problem 2
from collections import Counter

# Splitting the Words
words = input().lower().split(" ")
word_count = dict(Counter(words))

# Final Result and Print the Result
result = {word: count for word, count in word_count.items() if count > 1}
print(result)
```

```python
# Problem 3
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

for number in sensor_readings:
    if number % 2 == 0:
        hour = sensor_readings.index(number)
        valid_readings.append((hour, number))

print("Valid Sensor Readings (Hour, Value):", valid_readings)
```

```python
# Problem 4
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}

#Counting the Emails
for email in emails:
    domain = email.split("@")[1]   # get text after @
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)

# Checking the Percentage
for domain in domain_count:
    percentage = (domain_count[domain] / total_emails) * 100
    print(domain, ":", int(percentage), "%")
```

```python
# Problem 5
import numpy as np
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg = int(np.mean(sales))

for i in range(len(sales)):
    if(sales[i] > avg):
        print(f"Day {i + 1} : {sales[i]}")
```
